```python
import pandas as pd
import numpy as np

from IPython.display import display
```

```python
mca = pd.read_csv('marketing_customer_analysis.csv')

display(mca.head())
display(mca.tail())

mca.shape
```

```python
cols = []
for column in mca.columns:
    cols.append(column.lower())
mca.columns = cols

mca.head(30)
```

```python
mca.isna().sum()
```

```python
mca.isna().sum()
round(mca.isna().sum()/len(mca),4)*100
nulls_df = pd.DataFrame(round(mca.isna().sum()/len(mca),4)*100)
nulls_df
nulls_df = nulls_df.reset_index()
nulls_df
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
mca['vehicle type'].unique()
```

```python
mca_clean = mca.drop(['vehicle type'], axis=1)
mca_clean
```

```python
import statistics as stats
stats.mode(mca['state'])
```

```python
mca['state'] = mca['state'].fillna('California')
```

```python
mca['state'] = mca['state'].fillna(stats.mode(mca['state']))
```

```python
mca['response'] = mca['response'].fillna(stats.mode(mca['response']))
mca['months since last claim'] = mca['months since last claim'].fillna(stats.mode(mca['months since last claim']))
mca['number of open complaints'] = mca['number of open complaints'].fillna(stats.mode(mca['number of open complaints']))
mca['vehicle class'] = mca['vehicle class'].fillna(stats.mode(mca['vehicle class']))
mca['vehicle size'] = mca['vehicle size'].fillna(stats.mode(mca['vehicle size']))
```

```python
mca_clean['effective to date_new'] = mca_clean['effective to date']
mca_clean.head()
```

```python
mca_clean['effective to date_new'] = pd.to_datetime(mca_clean['effective to date_new'], errors='coerce')
mca_clean.dtypes
```

```python
mca_clean['effective to date_new'] = pd.DatetimeIndex(mca_clean['effective to date_new']).month
mca_clean.head()
```

```python
mcaX=mca_clean[mca_clean['effective to date_new']<4]
display(mcaX.head(20))
display(mcaX.tail(20))
```

```python
import pandas as pd
import numpy as np
```

```python
mca.describe(include=[np.object]).T
```

```python
summary = mca.describe().T
summary['iqr'] = summary['75%']-summary['25%']
summary['mean'] = list(map(lambda x: round(x,2), summary['mean']))

for col in summary.columns:
    summary[col] = summary[col].apply(lambda x : round(x, 2))
summary
```

```python
x, y, hue = "Sales Channel", "Response Proportion", "Response"
hue_order = ["No", "Yes"]

(data[x]
 .groupby(data[hue])
 .value_counts(normalize='True')
 .rename(y)
 .reset_index()
 .pipe((sns.barplot, 'data'), x=x, y=y, hue=hue))
```

```python
s1_2 = pd.crosstab(index=data['Sales Channel'],
                             columns=data['Response'],
                             normalize="index")
s1_2
```

```python
s1_2.plot(kind='bar', 
                    stacked=True, 
                    colormap='tab10', 
                    figsize=(10, 6))

plt.legend(loc="upper left", ncol=2)
plt.xlabel("Sales Channel")
plt.ylabel("Proportion of Responses")

plt.show()
```
