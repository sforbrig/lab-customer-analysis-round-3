```python
#Lab5:
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

lm = linear_model.LinearRegression()
```

```python
sns.scatterplot(x='Monthly Premium Auto',y='Total Claim Amount',data=mca_num)
plt.show()
```

```python
mca_num['Monthly Premium Auto'].corr(mca_num['Total Claim Amount'])
```

```python
y = mca_num['Total Claim Amount']
X = mca_num[['Monthly Premium Auto']] 
```

```python
lm.fit(X,y)
print(lm.intercept_)
print(lm.coef_)
```

```python
regression_line = lm.intercept_ + lm.coef_[0]*mca_num['Monthly Premium Auto']
plt.plot(mca_num['Monthly Premium Auto'], regression_line, c = 'red')
sns.scatterplot(x='Monthly Premium Auto',y='Total Claim Amount',data=mca_num)
plt.show()
```
